Task 1: Randomly generate a small dataset and show dimensionality reduction using PCA by writing the
code from scratch in python.

```python
import numpy as np
```

```python
# Step 1: Generate A Dummy DataSet
X = np.random.randint(10,100,100).reshape(10,10) 
X
```

```python
np.mean(X, axis=0)
```

```python
# Step 2: Mean Centering the data  
X_meaned = X - np.mean(X , axis = 0)
X_meaned
```

Data generated by the above code have dimensions (10,10) i.e. 10 examples and 10 attributes for each example. we calculated the mean of each attribute and subtracted that from every row of the respective column.



```python
# Step 3: Calculating the Covariance Matrix of the Mean-centered data.
cov_mat = np.cov(X_meaned , rowvar = False)
cov_mat
```

We can find easily calculate covariance Matrix using numpy.cov( ) method. The default value for rowvar is set to True, remember to set it to False to get the covariance matrix in the required dimensions.

```python
# Step 4: Calculating Eigenvalues and Eigenvectors of the covariance matrix
eigen_values , eigen_vectors = np.linalg.eigh(cov_mat)
print(eigen_values)
```

```python
eigen_vectors
```

NumPy linalg.eigh( ) method returns the eigenvalues and eigenvectors of a complex Hermitian or a real symmetric matrix.

```python
# Step 5: Sort the Eigenvalues in descending order

sorted_index = np.argsort(eigen_values)[::-1]
sorted_eigenvalue = eigen_values[sorted_index]

# Sort the eigenvectors 

sorted_eigenvectors = eigen_vectors[:,sorted_index]
```

np.argsort returns an array of indices of the same shape.



```python
# Step 6: Select a subset of the rearranged EigenValue matrix.
# Here, we select the first n eigenvectors, n is desired dimension of our final reduced data.
 
n_components = 4
eigenvector_subset = sorted_eigenvectors[:,0:n_components]
```

n_components = 4 means our final data should be reduced to just 4 variables. if we change it to 2 then we get our data reduced to 2 variables, so on.



```python
# Step 7: Transform the data 
X_reduced = np.dot(eigenvector_subset.transpose(),X_meaned.transpose()).transpose()
X_reduced
```

The final dimensions of X_reduced will be ( 10, 2 ) and originally the data was of higher dimensions ( 10, 10 ).

**Entire PCA Process In One Function**

```python
import numpy as np

def PCA(X , num_components):
     
    #Step-1
    X_meaned = X - np.mean(X , axis = 0)
     
    #Step-2
    cov_mat = np.cov(X_meaned , rowvar = False)
     
    #Step-3
    eigen_values , eigen_vectors = np.linalg.eigh(cov_mat)
     
    #Step-4
    sorted_index = np.argsort(eigen_values)[::-1]
    sorted_eigenvalue = eigen_values[sorted_index]
    sorted_eigenvectors = eigen_vectors[:,sorted_index]
     
    #Step-5
    eigenvector_subset = sorted_eigenvectors[:,0:num_components]
     
    #Step-6
    X_reduced = np.dot(eigenvector_subset.transpose() , X_meaned.transpose() ).transpose()
     
    return X_reduced
```

Task 3 : Use PCA code from sklearn.decomposition library and perform compression on your image.

```python
import matplotlib.image as mpimg
```

```python
from __future__ import print_function

import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_olivetti_faces
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = mpimg.imread('/content/pexels-photo-1987301.jpeg')
# Now, let's look at the size of this numpy array object img as well as plot it using imshow. 
print(img.shape)

plt.axis('off') 
plt.imshow(img) 
```

```python
#Okay, so the array has 674 rows each of pixel 1905. Let's reshape it into a format that PCA can understand. # 3600 = 1200 * 3 
img_r = np.reshape(img, (img.shape[0], 3*img.shape[1])) 
print(img_r.shape)
```

```python
# Great, now lets run RandomizedPCA with 64 components (8x8 pixels) and transform the image. 
ipca = PCA(64).fit(img_r) 
img_c = ipca.transform(img_r) 
print(img_c.shape)
print(ipca.explained_variance_ratio_)
print(np.sum(ipca.explained_variance_ratio_))
```

```python
#OK, now to visualize how PCA has performed this compression, let's inverse transform the PCA output and 
# reshape for visualization using imshow.
temp = ipca.inverse_transform(img_c) 
print(temp.shape) 
plt.imshow(temp)
```

```python
# reshaping 29355 back to the original 785 * 3 
temp = np.reshape(temp, (img.shape[0],img.shape[1],3)) 
print (temp.shape )
plt.imshow(temp)
```

Task 4 : Use PCA code from sklearn.decomposition library and perform dimensionality reduction on
olivetti_face dataset 

We will use the Olivetti Faces dataset. It has 10 faces each of 40 persons as 64x64 images.

```python
faces_data = fetch_olivetti_faces()
```

```python
n_samples, height, width = faces_data.images.shape
X = faces_data.data
n_features = X.shape[1]
y = faces_data.target
n_classes = int(max(y)+1)

print("Number of samples: {}, \nHeight of each image: {}, \nWidth of each image: {}, \nNumber of input features: {},\nNumber of output classes: {}\n".format(n_samples,height,
                                        width,n_features,n_classes))
```

Shuffle the data randomly and make train and test splits

```python
from sklearn.model_selection import train_test_split
```

```python
# Split into a training set (75%) and a test set (25%) 

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42)
```

```python
mean_image = np.mean(X_train,axis=0)
plt.figure
plt.imshow(mean_image.reshape((64,64)), cmap=plt.cm.gray)
plt.xticks(())
plt.yticks(())
plt.show()
```

Make a function for visualization of the images as an album

```python
def plot_gallery(images, h, w, titles=None, n_row=3, n_col=4):
    """
    Helper function to plot a gallery of portraits
    Taken from: http://scikit-learn.org/stable/auto_examples/applications/face_recognition.html
    """
    plt.figure(figsize=(1.8 * n_col, 2.4 * n_row))
    plt.subplots_adjust(bottom=0, left=.01, right=.99, top=.90, hspace=.35)
    for i in range(n_row * n_col):
        plt.subplot(n_row, n_col, i + 1)
        plt.imshow(images[i].reshape((h, w)), cmap=plt.cm.gray)
        if titles != None:
            plt.title(titles[i], size=12)
        plt.xticks(())
        plt.yticks(())
```

Visualize some faces from the training set

```python
chosen_images = X_train[:12]
chosen_labels = y_train[:12]
titles = ['Person #'+str(i) for i in chosen_labels]

plot_gallery(chosen_images, height, width, titles)
```

#Calculate a set of eigen-faces

We find the eigen vectors corresponding to the biggest eigen values of the covariance matrix of the data. These eigen vectors are the directions along which the data shows maximum amount of variation. Each eigen vector can be considered as an eigen face. We can represent any image in the dataset as a linear combination of these eigen faces with minimum error.

```python
#Reduce the dimensionality of the feature space
n_components = 150

#Finding the top n_components principal components in the data
pca = PCA(n_components=n_components, whiten=True).fit(X_train)

#Find the eigen-vectors of the feature space
eigenfaces = pca.components_.reshape((n_components, height, width))
```

#Visualize the eigen faces

```python
titles = ['eigen-face #'+str(i) for i in range(12)]
plot_gallery(eigenfaces, height, width, titles)
```

#Transform the data to the vector space spanned by the eigen faces

```python
#Projecting the data onto the eigenspace
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)

print("Current shape of input data matrix: ", X_train_pca.shape)
```
